```python
#df['Department'].fillna('NA',inplace=True)
#df['Job_Title'].fillna('NA',inplace=True)
#df['Performance_Rating'].fillna(df['Performance_Rating'].mean(),inplace=True)
#df['Experience_Years'].fillna(df['Experience_Years'].mean(),inplace=True)
#df['Salary_INR'].fillna(df["Salary_INR"].mean(),inplace=True)
#df['Hire_Date'] = df['Hire_Date'].fillna(pd.Timestamp.now().normalize())
#df['Location'].fillna('NA',inplace=True)
#df['Status'].fillna("Active",inplace=True)
#df['Work_Mode'].fillna('On-site',inplace=True)
df.fillna({
    'Department': 'NA',
    'Job_Title': 'NA',
    'Performance_Rating': df['Performance_Rating'].mean(),
    'Experience_Years': df['Experience_Years'].mean(),
    'Salary_INR': df['Salary_INR'].mean(),
    'Hire_Date': pd.Timestamp.now().normalize(),
    'Location': 'NA',
    'Status': 'Active',
    'Work_Mode': 'On-site'
}, inplace=True)
```

```python
# Convert 'Experience_Years' to numeric, coercing errors to NaN
df['Experience_Years'] = pd.to_numeric(df['Experience_Years'], errors='coerce')

# Fill NaN values with 0 (or another appropriate value like the mean) and convert to integer
df['Experience_Years'] = df['Experience_Years'].fillna(0).astype(int)

plt.figure(figsize=(12, 6))

sns.countplot(
    x='Experience_Years',
    data=df,
    order=sorted(df['Experience_Years'].unique()),
    palette='Set2'
)

plt.title('Employee Experience Distribution', fontsize=16)
plt.xlabel('Experience (Years)')
plt.ylabel('Number of Employees')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
